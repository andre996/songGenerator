```python
#Importing libreries
import lyricsgenius
import pandas as pd
import re
import math
```

```python
#Read CSV with songs'names
songs_list = pd.read_excel("listOfSongs.xlsx",header=0,index_col=0)
songs_list.head()
```

```python
genius = lyricsgenius.Genius('PcXy0z8QiTXUmvH1TBg2p15VEfTwjbYbbgIgPhcNVcWsmYE-qVOHAGOavd6NwRAi')
```

```python
#Clean songs
for key, value in songs_list.iterrows():
    try:
        lyric = genius.search_song(value['name'], value['Artist']).lyrics
        lyric = re.sub("[\(\[].*?[\)\]]", "",lyric)       
        lyric = lyric.replace("\n"," ")
        lyric = lyric.replace("]","")
        lyric = lyric.replace('"','')
    except:
        pass
    songs_list.loc[key,"lyric"] = songs_list.loc[key,"lyric"]
```
